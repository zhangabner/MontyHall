```python
from __future__ import print_function, division

%matplotlib inline

from thinkbayes2 import Pmf, Suite
import thinkplot
# from allendowney  http://allendowney.blogspot.com.au/

import math
import numpy as np
import pandas as pd
from scipy.special import gamma

class Euro(Suite):
    
    def Likelihood(self, data, hypo):
        """ 
        hypo is the prob of heads (0-100)
        data is a string, either 'H' or 'T'
        """
        x = hypo / 100
        if data == 'H':
            return x
        else:
            return 1-x   
        
euro1 = Euro(range(101))
evidence = 'H' * 12 + 'T' * 5  
for outcome in evidence:
    euro1.Update(outcome)
print (euro1.MaximumLikelihood())  #  only 3 H but 10 T left for Monty chooses

euro = Euro(range(101))

MH = Pmf()
MH[0]=1
MH[1]=1
MH.Normalize()

times=100
for j in range (0,times):
    N=30
    Winrandom = 0
    Winbayes = 0
    MNcase1=MH.Random()
    guessrandom = MH.Random()
    if MNcase1 == guessrandom:
        Winbayes = Winbayes +1
    if MNcase1 == guessrandom:
        Winrandom = Winrandom +1
    if MNcase1==0:
        euro.Update('H')
    if MNcase1==1:
        euro.Update('T')    
    
    for i in range (1,N):
        MNcase=MH.Random()
        if MNcase==0:
            euro.Update('H')
        if MNcase==1:
            euro.Update('T')    
        guessrandom = MH.Random()
        guessbayes = euro.MaximumLikelihood()
        if guessbayes <=50:
            guess =0
        else:
            guess =1
        if MNcase == guess:
             Winbayes = Winbayes +1
        if MNcase == guessrandom:
             Winrandom = Winrandom +1
    print ("first %d Winrandom %d Winbayes %d" % ( MNcase1,Winrandom,Winbayes))
```
